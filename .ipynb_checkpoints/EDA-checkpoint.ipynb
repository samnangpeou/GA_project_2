```python
train['SalePrice'] = train['SalePrice'].map(lambda x: np.log(x))
```

### Feature Engineering

```python
train = train[(np.abs(stats.zscore(train)) < 15).all(axis=1)]
#Deteching low z-score and getting rid of them.
```

```python
train['Lot Area'].describe()
```

```python
train['Overall Garage'] = train['Garage Area'] * train['Garage Cars']
train['Overall Great Fireplace'] = (train['Fireplace Qu_Ex'] + train['Fireplace Qu_Gd']) + train['Fireplaces']
train['TotalSF'] = train['2nd Flr SF'] + train['BsmtFin SF 1'] + train['BsmtFin SF 2'] + train['Bsmt Unf SF'] + train['1st Flr SF']
train['Recently Blt/Remod'] = train['Year Built'] + train['Year Remod/Add']
train['Bathrooms'] = train['Bsmt Half Bath'] + train['Full Bath'] + (train['Half Bath'] * .5) + train['Bsmt Full Bath']
train['Year Built/Remod'] = train['Year Built'] + train['Year Remod/Add']
train['Total Porch SF'] = train['Enclosed Porch'] + train['Screen Porch'] + train['Open Porch SF'] + train['3Ssn Porch'] + train['Wood Deck SF']
```

## Modeling!

```python
X = train[['TotalSF', 'Overall Qual', 'Gr Liv Area',
          'Overall Garage', 'Year Built/Remod', 'Bathrooms',
           'TotRms AbvGrd', 'Foundation_PConc',
          'Overall Great Fireplace', 'Exter Qual_Gd', 'Heating QC_Ex',
          'Total Porch SF']]
y = train['SalePrice']
lr = LinearRegression()
poly = PolynomialFeatures()
X_poly = poly.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_poly, y)
lr.fit(X_poly, y)
```

```python
cross_val_score(lr,X_train,y_train,cv=5)
```

```python
predictions = lr.predict(X_test)
```

```python
np.sqrt(mean_squared_error(np.exp(y_test),np.exp(predictions)))
```

```python
lr.score(X_train,y_train), lr.score(X_test,y_test)
```

### Trying a Ridge Model

```python
X_overfit = poly.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(
    X_overfit,
    y,
)
sc = StandardScaler()
Z_train = sc.fit_transform(X_train)        # NOTE: fit transform on X_train
Z_test = sc.transform(X_test)   
```

```python
ols = LinearRegression()
ols.fit(X_train,y_train)
```

```python
ridge_model = Ridge(alpha = 10)
# Fit.
ridge_model.fit(Z_train, y_train)
# Evaluate model using R2 bc this is a regression task
print(ridge_model.score(Z_train, y_train))
print(ridge_model.score(Z_test,y_test))
```

```python
r_alphas = np.logspace(0,5,100) #alphas range from 1 to 10^5

# Cross-validate over our list of ridge alphas.
ridge_cv = RidgeCV(alphas = r_alphas,
                  scoring = 'r2',
                  cv = 5)

# Fit model using best ridge alpha!
ridge_cv.fit(Z_train,y_train)
```

```python
ridge_model = Ridge(alpha = ridge_cv.alpha_)
# Fit.
ridge_model.fit(Z_train, y_train)
# Evaluate model using R2 b/c this is a regression task
print(ridge_model.score(Z_train, y_train))
print(ridge_model.score(Z_test, y_test))
```

```python
pred = ridge_model.predict(Z_train) 
np.sqrt(mean_squared_error(np.exp(y_train),np.exp(pred)))
```

### Let's try a LASSO

```python
# Set up a list of Lasso alphas to check.
l_alphas = np.logspace(-5, 2, 1000)

# Cross-validate over our list of Lasso alphas.
lasso_cv = LassoCV(alphas=l_alphas, cv=5, max_iter=50000, tol=0.1, selection='random', n_jobs=-1)

# Fit model using best ridge alpha!
lasso_cv.fit(Z_train, y_train);
```

```python
lasso_pred = lasso_cv.predict(Z_train)
```

```python
np.sqrt(mean_squared_error(np.exp(y_train),np.exp(lasso_pred)))
```
